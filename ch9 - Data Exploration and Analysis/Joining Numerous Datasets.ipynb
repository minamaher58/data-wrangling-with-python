```python
import xlrd
import agate
from xlrd.sheet import ctype_text
%store -r ranked
```

```python
cpi_workbook = xlrd.open_workbook('corruption_perception_index.xls')
cpi_sheet = cpi_workbook.sheets()[0]

for r in range(cpi_sheet.nrows):
    print(r, cpi_sheet.row_values(r))
```

```python
cpi_title_rows = zip(cpi_sheet.row_values(1), cpi_sheet.row_values(2))
cpi_titles = [t[0] + ' ' + t[1] for t in cpi_title_rows]
cpi_titles = [t.strip() for t in cpi_titles]
print(cpi_titles)
```

```python
cpi_rows = [cpi_sheet.row_values(r) for r in range(3, cpi_sheet.nrows)]
print(cpi_rows[0])
```

```python
def get_types(example_row):
    types = []
    
    for v in example_row:
        value_type = ctype_text[v.ctype]

        if value_type == 'text':
            types.append(agate.Text())
        elif value_type == 'number':
            types.append(agate.Number())
        elif value_type == 'xldate':
            types.append(agate.Date())
        else:
            types.append(agate.Text())
    return types


cpi_types = get_types(cpi_sheet.row(3))
```

```python
def get_table(new_arr, types, titles):
    try:
        table = agate.Table(new_arr, titles, types)
        return table
    except Exception as e:
        print(e)
```

```python
cpi_table = get_table(cpi_rows, cpi_types, cpi_titles)
print(cpi_table)
```

```python
print(cpi_titles)
```

```python
# solving the bad title warning
cpi_titles[0] = cpi_titles[0] + ' Duplicate'
cpi_table = get_table(cpi_rows, cpi_types, cpi_titles)
print(cpi_table)
```

```python
cpi_and_cl = cpi_table.join(ranked, 'Country / Territory', 'Countries and areas', inner=True)
cpi_and_cl.column_names
```

```python
# cpi is the 'Corruption Perceptions Index'
for r in cpi_and_cl.order_by('CPI 2013 Score').limit(10).rows:
    print('{}: {} - {}%'.format(r['Country / Territory'], r['CPI 2013 Score'], r['Total (%)']))
```

## Identifying Correlations

```python
# We want to determine whether perceived governemnt corruption and child labor rates are related.
import numpy

# this returns an error
numpy.corrcoef(cpi_and_cl.columns['Total (%)'].values(), cpi_and_cl.columns['CPI 2013 Score'].values())[0, 1]
```

```python
# converting Decimals to floats to avoid that error
numpy.corrcoef([float(t) for t in cpi_and_cl.columns['Total (%)'].values()], 
               [float(s) for s in cpi_and_cl.columns['CPI 2013 Score'].values()])[0, 1]
```

### our value of -0.36 indicates a weak correlation, but a correlation nevertheless. We can use this knowledge to dive deeper into datasets and what they mean.

## Identigying Outliers

```python
# we are going to use the agate table's standard deviation outlier method. This method returns a table of values at least three
# deviations above or below the mean.
```

```python
import agatestats
# agatestats.patch()
```

```python
std_dev_outliers = cpi_and_cl.stdev_outliers('Total (%)', deviations = 3, reject = False)
len(std_dev_outliers.rows)
```

```python
std_dev_outliers = cpi_and_cl.stdev_outliers('Total (%)', deviations = 5, reject = False)
len(std_dev_outliers.rows)
```

### this is not what we want. when we used five deviations, we did not see a change in the result. This is telling us our data is not very regularly distributed.

```python
mad = cpi_and_cl.mad_outliers('Total (%)')

for r in mad.rows:
    print(r['Country / Territory'], r['Total (%)'])
```
