```python
from datetime import datetime
from csv import reader
import pandas as pd
import pickle
import pprint
```

```python
# Load the list from the file
def load_zipped_data():
    with open('zipped_data.pkl', 'rb') as file:
        zipped_data = pickle.load(file)
        return zipped_data
```

```python
#checking if an important value is missing in the first row
zipped_data = load_zipped_data()
for answer in zipped_data[0]:
    if not answer[1]:
        print(answer)
```

```python
# looping over every row
zipped_data = load_zipped_data()
for row in zipped_data:
    for key, val in row:
        if (val is None) or (not val):
            print(val)
```

```python
zipped_data = load_zipped_data()

na_count = {}
for row in zipped_data:
    for resp in row:
        question = resp[0][1]
        answer = resp[1]
        if answer == 'NA':
            if question in na_count.keys():
                na_count[question] += 1
            else:
                na_count[question] = 1
                
pprint.pprint(na_count)
```

```python
# exploring the distribution of types in our responses
zipped_data = load_zipped_data()

datatypes = {}

start_dict = {
    'digit': 0, 'boolean': 0,
    'empty': 0, 'time_related':0,
    'text': 0, 'unknown': 0
}

for row in zipped_data:
    for resp in row:
        question = resp[0][1]
        answer = resp[1]
        key = 'unknown'
        if answer.isdigit():
            key = 'digit'
        elif answer in ['Yes', 'No', 'True', 'False']:
            key = 'boolean'
        elif answer.isspace():
            key = 'empty'
        elif answer.find('/') > 0 or answer.find(':') > 0:
            key = 'time_related'
        elif answer.isalpha():
            key = 'text'
        if question not in datatypes.keys():
            datatypes[question] = start_dict.copy()
        
        datatypes[question][key] += 1
                
pprint.pprint(datatypes)
```
