```python
from datetime import datetime
from csv import reader
import pandas as pd
import pickle
```

```python
# Load the list from the file
with open('zipped_data.pkl', 'rb') as file:
    zipped_data = pickle.load(file)
    zipped_data = list(zipped_data[0])
```

### formatting Strings

```python
for x in zipped_data:
    print('Question: {}\nAnswer: {}'.format(x[0], x[1]))
```

```python
for x in zipped_data:
    print('Question: {[1]}\nAnswer: {}'.format(x[0], x[1]))
```

### formatting numbers

```python
# Example of format with number
example_dict = {
    'float_number': 1324.321325493,
    'very_large_integer': 43890923148390283,
    'percentage': .324
}

string_to_print = "float: {float_number:.4f}\n"
string_to_print += "integer: {very_large_integer:,}\n"
string_to_print += "percentage: {percentage:.2%}"

print(string_to_print.format(**example_dict))
```

### formatting dates

```python
for x in enumerate(zipped_data[0:20]):
    print(x)
```

```python
start_string = '{}/{}/{} {}:{}'.format(
    zipped_data[8][1], zipped_data[7][1], zipped_data[9][1],
    zipped_data[13][1], zipped_data[14][1]
)

print(start_string)
```

```python
start_time = datetime.strptime(start_string, '%m/%d/%Y %H:%M')
print(start_time)
```

```python
# the order datetime expects data in is year,month,day,hour,minute
end_time = datetime(
    int(zipped_data[9][1]), int(zipped_data[8][1]),
    int(zipped_data[7][1]), int(zipped_data[15][1]),
    int(zipped_data[16][1]),
)

print(end_time)
```

```python
# doing math with date time
duration = end_time - start_time
print(duration)
print(duration.days)
print(duration.total_seconds())
minutes = duration.total_seconds() / 60
print(minutes)
```

```python
# convert datetime objects back into formatted strings
print(end_time.strftime('%m/%d/%Y %H:%M:%S'))
print(start_time.ctime())
print(start_time.strftime('%Y-%m-%dT%H:%M:%S'))
```
